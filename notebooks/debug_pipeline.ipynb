# SynthMorph Debug Pipeline Visualization

This notebook visualizes the synthetic generation pipeline for either `sm-shapes` or `sm-brains`, plotting:
- Base generated structural labels
- Derived Deformation fields (Displacements)
- Resulting Synthesized MRI Image Pairs (Moving & Fixed)


```python
import sys
import os
sys.path.append('..') # Root module inclusion

import torch
import matplotlib.pyplot as plt
from src.pipeline.sm_shapes_pipeline import SynthMorphShapesPipeline
from src.pipeline.sm_brains_pipeline import SynthMorphBrainsPipeline
```

```python
# Initialize shapes pipeline natively determining CPU / GPU from your env
shapes_pipeline = SynthMorphShapesPipeline()

print("Generating image pair...")
with torch.no_grad():
    # Extract a batch size 1 pairing
    m, f, s_m, s_f = shapes_pipeline.generate_pair(batch_size=1)
```

```python
def plot_middle_slices(m_t, f_t, sm_t, sf_t):
    # Isolate center slice across depth/Z axis
    D = m_t.shape[2]
    mid = D // 2
    
    m_slice = m_t[0, 0, mid, :, :].cpu().numpy()
    f_slice = f_t[0, 0, mid, :, :].cpu().numpy()
    
    sm_slice = sm_t[0, 0, mid, :, :].cpu().numpy()
    sf_slice = sf_t[0, 0, mid, :, :].cpu().numpy()
    
    fig, axes = plt.subplots(2, 2, figsize=(10, 10))
    axes[0, 0].imshow(sm_slice, cmap='tab20')
    axes[0, 0].set_title("Moving Label Map (Shapes)")
    
    axes[0, 1].imshow(sf_slice, cmap='tab20')
    axes[0, 1].set_title("Fixed Label Map (Shapes)")
    
    axes[1, 0].imshow(m_slice, cmap='gray')
    axes[1, 0].set_title("Moving MRI")
    
    axes[1, 1].imshow(f_slice, cmap='gray')
    axes[1, 1].set_title("Fixed MRI")
    
    plt.tight_layout()
    plt.show()
    
plot_middle_slices(m, f, s_m, s_f)
```
